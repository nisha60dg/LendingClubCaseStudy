Creating plots to Analyze Data

```python
loan_status_counts = df['loan_status'].value_counts()
```

```python
loan_status_counts
```

```python
# Plot distribution of loan status
plt.figure(figsize=(8, 5))
loan_status_counts.plot(kind='bar', color='skyblue')
plt.title("Loan Status Distribution")
plt.ylabel("Count")
plt.xlabel("Loan Status")
plt.show()
```

```python
#This bar chart showing how many loans are in default versus fully paid. 
```

```python
# Filter defaulters
defaulters = df[df['loan_status'] == 'Charged Off']
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='term', hue='loan_status', data=df)
plt.title("Default Rate by Term")
plt.xlabel("Loan Term")
plt.ylabel("Count")
plt.show()
```

# BIVARIATE ANALYSIS

```python
# Analyze default rate by loan amount
plt.figure(figsize=(8, 5))
sns.boxplot(x='loan_status', y='loan_amnt', data=df)
plt.title("Default Rate by Loan Amount")
plt.xlabel("Loan Status")
plt.ylabel("Loan Amount")
plt.show()
```

```python
#The box plot showing if loans are larger then defaulter rate also increases.
```

```python
# Analyze default rate by loan purpose
plt.figure(figsize=(12, 6))
sns.countplot(x='purpose', hue='loan_status', data=df, palette='viridis')
plt.title("Loan Status by Purpose")
plt.xlabel("Loan Purpose")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Loan Status")
plt.show()
```

```python
# 1. Loan Purpose vs Loan Status
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='purpose', hue='loan_status', order=df['purpose'].value_counts().index)
plt.title("Loan Purpose vs Loan Status")
plt.xlabel("Loan Purpose")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Loan Status", loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.catplot(x="loan_status", y="purpose",kind="bar",data=df);
plt.show()
```

```python
# 2. Home Ownership vs Loan Status
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='home_ownership', hue='loan_status', order=df['home_ownership'].value_counts().index)
plt.title("Home Ownership vs Loan Status")
plt.xlabel("Home Ownership")
plt.ylabel("Count")
plt.legend(title="Loan Status", loc='upper right')
plt.show()
```

```python
df['int_rate'] = df['int_rate'].str.replace('%', '', regex=False).astype(float)
```

```python
# 3. Interest Rate Distribution for Defaulters vs Non-defaulters
plt.figure(figsize=(8, 5))
sns.kdeplot(df[df['loan_status'] == 'Charged Off']['int_rate'], label='Defaulters', shade=True)
sns.kdeplot(df[df['loan_status'] != 'Charged Off']['int_rate'], label='Non-Defaulters', shade=True)
plt.title("Interest Rate Distribution")
plt.xlabel("Interest Rate (%)")
plt.legend()
plt.show()
```

```python
loan_status_counts.to_csv('loan_status_distribution.csv', index=True)
occupation_defaulters.to_csv('occupation_defaulters.csv', index=True)
purpose_defaulters.to_csv('purpose_defaulters.csv', index=True)
```

```python
# Loan Status Pie Chart
plt.figure(figsize=(8, 8))
df['loan_status'].value_counts().plot.pie(
    autopct='%1.1f%%', startangle=90, explode=[0.1] + [0] * (df['loan_status'].nunique() - 1), 
    colors=sns.color_palette('pastel'), shadow=True
)
plt.title("Loan Status Distribution")
plt.ylabel("")  # Hides the y-label
plt.show()
```

```python
# Loan Amount Histogram
plt.figure(figsize=(8, 5))
plt.hist(df['loan_amnt'], bins=30, color='skyblue', edgecolor='black')
plt.title("Loan Amount Distribution")
plt.xlabel("Loan Amount ($)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter Plot: Loan Amount vs Annual Income
plt.figure(figsize=(8, 5))
plt.scatter(df['annual_inc'], df['loan_amnt'], alpha=0.5, color='teal')
plt.title("Loan Amount vs Annual Income")
plt.xlabel("Annual Income ($)")
plt.ylabel("Loan Amount ($)")
plt.grid(True)
plt.show()
```

```python
# Bubble Plot: Loan Amount by Home Ownership
home_ownership_counts = df.groupby('home_ownership')['loan_amnt'].sum()
plt.figure(figsize=(8, 5))
plt.scatter(home_ownership_counts.index, home_ownership_counts.values, 
            s=home_ownership_counts.values / 1000, alpha=0.6, color='orange', edgecolors='black')
plt.title("Total Loan Amount by Home Ownership")
plt.xlabel("Home Ownership")
plt.ylabel("Total Loan Amount ($)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Pair Plot: Relationship Between Numerical Variables
sns.pairplot(df[['loan_amnt', 'annual_inc', 'int_rate', 'dti']], diag_kind='kde', corner=True)
plt.show()
```

```python
# Heatmap: Correlation of Numerical Features
plt.figure(figsize=(10, 8))
sns.heatmap(df[['loan_amnt', 'annual_inc', 'int_rate', 'dti']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```
